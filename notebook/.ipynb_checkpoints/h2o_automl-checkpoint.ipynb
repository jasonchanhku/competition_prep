# Data Preparation

* train test split of 80:20
* add daate features

```python
def create_features(df, label=None):
    """
    Creates time series features from datetime index
    """
    df = df.copy()
    df['date'] = pd.to_datetime(df.index)
    df['hour'] = df['date'].dt.hour
    df['dayofweek'] = df['date'].dt.dayofweek
    df['quarter'] = df['date'].dt.quarter
    df['month'] = df['date'].dt.month
    df['year'] = df['date'].dt.year
    df['dayofyear'] = df['date'].dt.dayofyear
    df['dayofmonth'] = df['date'].dt.day
    df['weekofyear'] = df['date'].dt.weekofyear
    df.set_index('date', inplace=True)
    
    return df
```

```python
df = create_features(df)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df_train = df.iloc[:int(df.shape[0]*0.8), :]
df_test = df.iloc[int(df.shape[0]*0.8): , :]
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df_train.index,df_train['pollution'])
plt.plot(df_test.index,df_test['pollution'])
plt.ylabel('pollution',fontsize=18)
plt.legend(['train','test'])
plt.title('Train / Test Split', fontsize=30)
plt.show()
```

# Initialize H2O Session

```python
h2o.init(nthreads=-1)
```

## Convert to H2O Frame

```python
hf_train = h2o.H2OFrame(df_train)
hf_test = h2o.H2OFrame(df_test)
```

```python
hf_train.describe()
```

```python
X = hf_train.columns[1:]
y = hf_train.columns[0]
```

```python
X
```

```python
y
```

```python
900/60
```

```python
aml = H2OAutoML(max_runtime_secs=900, seed=42)
```

```python
aml.train(x=X, y=y, training_frame=hf_train, leaderboard_frame=hf_test)
```

```python
lb = aml.leaderboard
lb
```

```python
leader_model = aml.leader
```

```python
hf_test_predict = leader_model.predict(hf_test)
```

```python
df_results = pd.DataFrame()
df_results['ground_truth'] = df_test['pollution'].reset_index(drop=True)
df_results['predictions'] = h2o.as_list(hf_test_predict,use_pandas=True)
df_results.head()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(df_results.index,df_results['ground_truth'])
plt.plot(df_results.index,df_results['predictions'])
plt.ylabel('pollution',fontsize=18)
plt.legend(['actual','predict'])
plt.title('Actual vs Predict', fontsize=30)
plt.show()
```

```python
def process_data(train , test, seq_len):
    
    train.drop(['wnd_dir'], axis=1, inplace=True)
    test.drop(['wnd_dir'], axis=1, inplace=True)
    
    num_vars = len(train.columns) - 1
    print(f"will shape to {vars} variables")
    
    X_train = []
    y_train = []
    
    for i in range(seq_len, len(train)):
        X_train.append(train.values[i-seq_len:i, 1:])
        y_train.append(train.values[i, 0])
    
    # to make sure test has the 60 values before
    df_total = pd.concat([train, test], axis = 0)
    inputs = df_total.values[len(df_total) - len(test)-60: , :]
    
    X_test = []
    y_test = []
    
    for i in range(seq_len, len(inputs)):
        X_test.append(inputs.values[i-seq_len:i, 1:])
        y_test.append(inputs.values[i, 0])
        
    X_train, y_train, X_test, y_test = np.array(X_train).reshape(-1, seq_len, num_vars), 
    np.array(y_train).reshape(-1, 1), np.array(X_test).reshape(-1, seq_len, num_vars), np.array(y_test).reshape(-1, 1)
    
    
    
    return X_train, y_train, X_test, y_test
```

```python
X_train, y_train, X_test, y_test = process_data(df_train, df_test, 60)
```
